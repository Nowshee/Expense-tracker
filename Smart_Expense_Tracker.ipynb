```python
class Expense:
    def __init__(self,category,amount,date):
        self.category = category
        self.amount = amount
        self.date = date 
    def __str__(self):
        return f"{self.date} | {self.category} | ₹{self.amount}"   # ctrl Alt 4 for rupee symbol
print("Welcome to Smart Expense Tracker")
print("Menu")
lst = []
def add_expense():
    print("Enter your Expenses")
    category = input("Enter Category")
    amount = input("Enter Amount")
    date = input("Enter Date")
    e = Expense(category,amount,date)
    lst.append(e)
    with open ("expense.txt",'a',encoding = "utf-8") as f:
        for line in lst:
            f.write(str(line)+"\n")
def view_expense():
    with open ("expense.txt",'r',encoding = "utf-8") as f:
        text = f.read()
        if text.strip() == "":
            print("No Expenses Recorded Yet")
        else:
            print("Date | Category | Amount")
            print(text)
def show_total():
    total = 0
    with open ("expense.txt",'r',encoding = "utf-8") as f:
        while True:
            lines = f.readline()
            if not lines:
                break
            m= lines.split("|")[2]
            amount = int(m.replace("₹",""))
            total += amount
        print(f"Total = ₹{total}")   
while True:
    print("Press 1 to Add Expenses\nPress 2 to View All Expenses\nPress 3 to Show Total\nPress 4 to Exit ")
    user = int(input())
    try:
        if user < 1 or user > 4:
            raise ValueError("Invalid Input")
    except:
        print("You must Enter a Valid Input to Continue")
    if user == 1:
        add_expense()   
    if user == 2:
        view_expense()
    if user == 3:
        show_total()   
    if user == 4:
        print("Thank you for using our App")
        break

```
